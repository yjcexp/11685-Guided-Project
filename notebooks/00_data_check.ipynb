# 00 - Data Check

This notebook is used for checking data integrity and understanding data structure.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Add src to path
import sys
sys.path.append('..')
from src.data_utils import load_metadata, load_eeg_file, normalize_eeg
```

```python
# Load metadata
metadata = load_metadata('data/processed/metadata.csv')
print(f"Total samples: {len(metadata)}")
metadata.head()
```

```python
# Check class distribution
if 'class_label' in metadata.columns:
    class_counts = metadata['class_label'].value_counts().sort_index()
    print("Class distribution:")
    print(class_counts)
    
    plt.figure(figsize=(10, 5))
    class_counts.plot(kind='bar')
    plt.title('Class Distribution')
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.show()
```

```python
# Load and inspect a sample EEG signal
sample_row = metadata.iloc[0]
eeg_path = sample_row['eeg_path']
eeg_signal = load_eeg_file(eeg_path)

print(f"EEG shape: {eeg_signal.shape}")
print(f"EEG dtype: {eeg_signal.dtype}")
print(f"EEG range: [{eeg_signal.min():.3f}, {eeg_signal.max():.3f}]")
```

```python
# Visualize EEG signal
plt.figure(figsize=(15, 4))
plt.plot(eeg_signal.flatten())
plt.title('Sample EEG Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Check normalization
eeg_normalized = normalize_eeg(eeg_signal, method='zscore')
print(f"Normalized EEG range: [{eeg_normalized.min():.3f}, {eeg_normalized.max():.3f}]")
print(f"Normalized EEG mean: {eeg_normalized.mean():.3f}, std: {eeg_normalized.std():.3f}")
```

## Summary

This notebook helps verify:
- Metadata is correctly loaded
- Class distribution is reasonable
- EEG signals have expected shape and range
- Normalization works as expected
