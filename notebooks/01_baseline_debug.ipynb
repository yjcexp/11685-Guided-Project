```python
# Test BaselineMLP
# Create dummy input
batch_size = 4
input_dim = 1000
dummy_input = torch.randn(batch_size, input_dim).to(device)

model = BaselineMLP(
    input_dim=input_dim,
    hidden_dims=[256, 128],
    num_classes=10
).to(device)

output = model(dummy_input)
print(f"Input shape: {dummy_input.shape}")
print(f"Output shape: {output.shape}")
print(f"Output range: [{output.min():.3f}, {output.max():.3f}]")
```

```python
# Test EEGCNN
batch_size = 4
channels = 1
time_steps = 1000
dummy_input = torch.randn(batch_size, channels, time_steps).to(device)

model = EEGCNN(
    in_channels=channels,
    conv_channels=[32, 64, 128],
    kernel_size=3,
    pool_size=2,
    num_classes=10
).to(device)

output = model(dummy_input)
print(f"Input shape: {dummy_input.shape}")
print(f"Output shape: {output.shape}")
print(f"Output range: [{output.min():.3f}, {output.max():.3f}]")
```

```python
# Count parameters
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

mlp = BaselineMLP(input_dim=1000, hidden_dims=[256, 128], num_classes=10)
cnn = EEGCNN(in_channels=1, conv_channels=[32, 64, 128], kernel_size=3, pool_size=2, num_classes=10)

print(f"MLP parameters: {count_parameters(mlp):,}")
print(f"CNN parameters: {count_parameters(cnn):,}")
```
